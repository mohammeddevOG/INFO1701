```python
def calcUSPopulation(startPopulation):

    # So we Are trying to find the USPopulation in one year
    # By using the rate of change we are givien: Birth,Death, and Imigrants

    # We have to return after calculating the new population

    secondInAYear = 365*24*60*60

    # The Increase Per Year
    birthPerYear = secondInAYear//9
    deathPerYear = secondInAYear//10
    immigrantsPerYear = secondInAYear//47


    # Now we do the New Population
    populationPerYear = startPopulation + birthPerYear - deathPerYear + immigrantsPerYear

    # returning the New Population
    return populationPerYear
```
